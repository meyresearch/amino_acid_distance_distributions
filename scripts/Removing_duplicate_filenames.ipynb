```python
import numpy as np
import glob
import pandas as pd
import matplotlib.pyplot as plt
```

```python
files = glob.glob('/home/jguven/Projects/sequence_distance_distribution/data/alphafold/pdb_files/*.pdb')
```

```python
df = pd.read_csv('/home/amey/secondary_structures_300.csv').dropna()
```

```python
protein_name = []
for pdb in df['filename']:
    with open(pdb) as f:
        head = [next(f) for x in range(2)]
        protein_name.append(head[-1].strip().split('FOR')[-1])
```

```python
df['protein_name'] = protein_name
```

```python
new_df = df.drop_duplicates(subset='protein_name')
```

```python
pd.DataFrame.to_csv(new_df, 'unique_secondary_structures_300.csv')
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.subplots(1,1)

ax.scatter(df["H"], df["beta"], alpha=0.5)
ax.scatter(new_df["H"], new_df["beta"])
ax.set_xlabel(r"$\alpha$-helix %")
ax.set_ylabel(r"$\beta$-sheet %")
```
