```python
sec_df = pd.read_csv("../data/alphafold/secondary_structures_102.csv").dropna()
```

```python
conf_df = pd.read_csv("../data/alphafold/confidences_100.csv")
```

```python
conf_df
```

```python
sec_df["filename"] = conf_df["filename"]
sec_df["mean_conf"] = conf_df["mean_conf"]
sec_df["std_conf"] = conf_df["std_conf"]
```

```python
plt.figure(figsize=(6,6))
plt.scatter(sec_df['H'],sec_df['B']+sec_df['E'], c=sec_df['mean_conf'], cmap='coolwarm_r')
plt.xlabel('helix_percent')
plt.ylabel('beta_percent')
```

```python
secondary_df = sec_df[["H","B","E","mean_conf","std_conf","filename"]]
```

```python
secondary_df
```

```python
secondary_df["beta"] = secondary_df["B"] + secondary_df["E"]
```

```python
alphas_df = secondary_df[secondary_df["H"] > 0.4]
alphas_df = alphas_df.drop(columns={"B", "E", "beta"})
alphas_df
```
